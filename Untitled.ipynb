# Trend Analysis (Inflation)


```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of countries to include in the plot
countries = ["India", "China", "Bangladesh", "Indonesia", "Japan", "Bhutan", "Pakistan", "Russia", "Saudi Arabia"]

# Filter data for inflation, average consumer prices, and Percent Change in Units
inflation_data = df_trend[
    (df_trend["Country"].isin(countries)) & 
    (df_trend["Subject Descriptor"] == "Inflation, average consumer prices") & 
    (df_trend["Units"] == "Percent change")
]
nrows = 3
ncols = 3
# Create a figure with subplots 
fig, axes = plt.subplots(nrows=nrows, ncols= ncols, figsize=(ncols*5, nrows*5))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Plot each country's inflation trend in a separate subplot
for i, country in enumerate(countries):
    country_data = inflation_data[inflation_data["Country"] == country]
    
    sns.lineplot(data=country_data, x="Year", y="Value", ax=axes[i], marker="o")
    
    # Set the title, labels, and grid for each subplot
    axes[i].set_title(f"{country} - Inflation Trend", fontsize=10)
    axes[i].set_xlabel("Year", fontsize=8)
    axes[i].set_ylabel("Inflation (%)", fontsize=8)
    axes[i].grid(True)
    axes[i].tick_params(axis='x', rotation=45, labelsize=8)
    axes[i].tick_params(axis='y', labelsize=8)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

# Finding Fiscal Defecit of Major Asian Countries

### fiscal defecit a very interesting topic in economics, it shows how much a country is generating revenue and how much it is spending like if a country is spending more and generating less revenue then it will have to take loan to fulfill its demand and certainly because of that factor its fiscal defecit will be higher

```python
# Filter relevant columns (adjust as needed to match your dataset)
revenue_data = df_trend[
    (df_trend["Subject Descriptor"] == "General government revenue") &
    (df_trend["Country"].isin(countries))  
]

expenditure_data = df_trend[
    (df_trend["Subject Descriptor"] == "General government total expenditure") &
    (df_trend["Country"].isin(countries))  
]

# Merge revenue and expenditure data
fiscal_df = pd.merge(revenue_data, expenditure_data, on=["Country", "Year"], suffixes=("_revenue", "_expenditure"))

# Calculate fiscal deficit as expenditure minus revenue
fiscal_df["Fiscal Deficit"] = fiscal_df["Value_expenditure"] - fiscal_df["Value_revenue"]
```

```python
plt.figure(figsize=(14, 8))

sns.lineplot(data=fiscal_df, x="Year", y="Fiscal Deficit", hue="Country", marker="o", errorbar=None)

# Adding labels and title
plt.title("Fiscal Deficit Trend (2024) for Selected Asian Countries")
plt.xlabel("Year")
plt.ylabel("Fiscal Deficit (% of GDP)")
plt.xticks(rotation=45)
plt.grid(True)

# Display the legend with the countries
plt.legend(title="Countries")
plt.tight_layout()

# Show the plot
plt.show()
```

# Correlation between Inflation, Exports and Investment

### Here I have created a heatmap of three indicators, so it will basically show how one indicator can affect another indicator:
### like rise in infaltio will lead to lesser investment in any country.

```python
# Filter relevant data for investment, exports, and inflation
investment_data = df_trend[
    (df_trend["Subject Descriptor"] == "Total investment") &
    (df_trend["Country"].isin(countries)) 
]

exports_data = df_trend[
    (df_trend["Subject Descriptor"] == "Volume of exports of goods") &
    (df_trend["Country"].isin(countries))
]

inflation_data = df_trend[
    (df_trend["Subject Descriptor"] == "Inflation, average consumer prices") &
    (df_trend["Units"] == "Percent change") & 
    (df_trend["Country"].isin(countries))
]

# Merge all three datasets on Country and Year
merged_data = pd.merge(investment_data, exports_data, on=["Country", "Year"], suffixes=("_investment", "_exports"))
merged_data = pd.merge(merged_data, inflation_data, on=["Country", "Year"])

# Rename the columns for better clarity
merged_data.rename(columns={
    "Value_investment": "Investment",
    "Value_exports": "Exports",
    "Value": "Inflation"
}, inplace=True)

# Select relevant columns for correlation calculation
corr_data = merged_data[["Year", "Country", "Investment", "Exports", "Inflation"]]

# Calculate the correlation matrix
correlation_matrix = corr_data[["Investment", "Exports", "Inflation"]].corr()

# Plotting the heatmap for correlation
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True, square=True)
plt.title("Correlation Matrix for Investment, Exports, and Inflation")
plt.show()
# for country in countries:
#     country_data = corr_data[corr_data["Country"] == country]
#     corr_matrix = country_data[["Investment", "Exports", "Inflation"]].corr()
    
#     plt.figure(figsize=(6, 4))
#     sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", square=True)
#     plt.title(f"Correlation - {country}")
#     plt.tight_layout()
#     plt.show()
```

# Clustering of Different Economies using K-means

### here I have did clustering basically a unsupervised learning and under this we will use K-means machine learning algorithm and K-Means is basically used to group similar data points into the cluster.
### Like here i have bunch of data containing several indicators so here K-means will group countries with similar Indicators and it will keep first similar countries in one cluster, then other in second one and so on.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Choose latest year (you can adjust to any specific year like 2023 or 2022)
latest_year = 2024

# Indicators to include in clustering
cluster_indicators = [
    "Total investment",
    "Inflation, average consumer prices",
    "Volume of exports of goods",
    "Unemployment rate",
    "General government revenue",
    "General government total expenditure"
]

# Filter and pivot data
df_cluster = df_trend[
    (df_trend["Year"] == latest_year) &
    (df_trend["Subject Descriptor"].isin(cluster_indicators)) &
    (df_trend["Units"].isin(["Percent of GDP", "Percent change"]))  # choose % values for consistency
]

# Pivot to wide format
df_pivot = df_cluster.pivot_table(
    index="Country",
    columns="Subject Descriptor",
    values="Value",
    aggfunc="mean"  # in case of duplicates
).dropna()

# Standardize the data for clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pivot)

# Elbow method to find optimal clusters
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 10), inertia, marker='o')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
# Apply KMeans with optimal clusters (e.g., 3)
optimal_k = 3  # Change this if your elbow plot suggests a different number
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_pivot["Cluster"] = kmeans.fit_predict(X_scaled)

# Reset index to have Country as column
df_clusters = df_pivot.reset_index()

# Display the cluster assignment
print(df_clusters[["Country", "Cluster"]].sort_values("Cluster"))
```

# Forecasting Inflation rate of India from 2025 to 2029

```python
from sklearn.linear_model import LinearRegression
import numpy as np


# Filter trend data
country = 'India'
indicator = 'Inflation, average consumer prices'

df_filtered = df_trend[(df_trend['Country'] == country) &
                       (df_trend['Units']== "Percent change") & 
                       (df_trend['Subject Descriptor'] == indicator)].dropna()

# Ensure data is sorted
df_filtered = df_filtered.sort_values(by='Year')
```

### I have used linear regression here for forecasting because it is the basic one and its my first step toward this type of project so I first learnt it and then applied it. So in this its basically train on past years anf fits an equation, Value = m*year +c, It has certain Disadvantages like its accuracy can be low if there is sudden spikes and dips in the graph to improve result we can use Polynomial regression(for curvy trend), ARIMA or prophet model(better for time series)


```python
# Independent (X) and dependent (y) variables
X = df_filtered['Year'].values.reshape(-1, 1)
y = df_filtered['Value'].values

# Create and train model
model = LinearRegression()
model.fit(X, y)
```

```python
# Forecast for next 5 years
future_years = np.arange(df_filtered['Year'].max() + 1, df_filtered['Year'].max() + 6).reshape(-1, 1)
future_preds = model.predict(future_years)

# Create DataFrame of predictions
forecast_df = pd.DataFrame({
    'Year': future_years.flatten(),
    'Forecasted Value': future_preds
})

print(forecast_df)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_filtered['Year'], df_filtered['Value'], label='Actual', marker='o')
plt.plot(forecast_df['Year'], forecast_df['Forecasted Value'], label='Forecast', marker='x', linestyle='--')
plt.title(f"{indicator} Forecast for {country}")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
